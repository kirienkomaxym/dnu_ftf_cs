## Списки | Lists

Списки - це структура даних, яка дозволяє хранити інші структури даних в собі, та дозволяє проводити операції над ними. В інших мовах програмування така структура даних зазвичай називається масив (array). В пайтоні цей тип даних має тип `list`. Лист створюється за допомогою квадратних скобок на початку `[` та в кінці `]`, кожен елемент має бути відокремлений комою `,`.

Тобто загальне створення ліста - `[elem1, elem2, elem3]`, для створення змінної відповідно `mylist = [elem1, elem2, ... elemN]`

```python
primes = [2, 3, 5, 7, 11, 13] # int list
rainbow = ['Red', 'Orange', 'Yellow', 'Green', 'Blue', 'Indigo', 'Violet']  # str list
bools = [True, False, True, False]  # bool list
```

```python
mixed_list = [1, "string", True, 0.5, 1e-4]  # mixed
natural_digits = [1, 2, 3, 4, 5, 6, 7, 8, 9]
```

```python
my_list = [1, 2, 3]
print(type(my_list))

empty_list = []
print(type(empty_list))
```

```python
sub_list = [1, 2, 3]
sub_sub_list = [["a", "b", "c"], 123]
my_list = [sub_list, sub_sub_list]
print(f"{my_list=}")
```

Альтернативою створюванню листів є функція `list` (вона ж і тип даних), за допомогою якої можна створити лист, або перетворити якусь послідовніть на ліст:

```python
empty_list = list()
print(f"{empty_list = }")
my_list = list("abcd")
print(f"{my_list = }")
```

Тобто, створення листа `a = list(1, 2, 3, 4, 5)` є неможливим

```python
a = list(1, 2, 3, 4, 5)
```

В першу чергу, листи корисні для збереження даних, додавання, віднімання, та розличних операцій над ними. Операції над даними в листі можуть бути виконані як завдяки операторам - `+, -, ==`, функціям - `bool(my_list)` `type(my_list`, і методам типу даних `list`. Також лісти підтримують вже знайому вам індексацію та можливість взяти зріз від послідовності. На відміну від вже знайомих вам строк `str`, лісти підтримують зміну даних в послідовності за допомогою індексації.

### Методи

Як це було зі строками, тип даних `list` також має власні методи, розберемо деякі з них:

```python
# append
my_list = []
print(f"{my_list = }")
my_list.append("Hello!")
print(f"{my_list = }")
my_list.append(["A List", "With", "Some", "Variables"])
print(f"{my_list = }")

# notice - that you don't have to create a new variable
appended = my_list.append("Hello #2!")
print(f"{appended = }")
print(f"{type(appended) = }")
```

```python
# extend
my_list = []
print(f"{my_list = }")
my_list.extend("Hello!")
print(f"{my_list = }")
my_list.extend(["A List", "With", "Some", "Variables"])
print(f"{my_list = }")
```

```python
# index
my_list = [1, 2, 3, 4, 5]
index = my_list.index(5)
print(f"{index = }")
print(f"{my_list[index] = }")

index2 = my_list.index(6)
print(f"{index = }")
```

```python
# count
my_list = [11, 22, 1, 1, 1, 2, 2, 2]
n = my_list.count(1)
print(f"{n = }")
k = my_list.count(True)
print(f"{k = }")
f = my_list.count(42)
print(f"{f = }")
```

```python
# pop
my_list = [1, 2, 3, 4, 5]
print(f"{my_list = }")
popped = my_list.pop()  # no arg specified

print(f"{popped = }")
print(f"{my_list = }")

popped2 = my_list.pop(2)  # no arg specified
print(f"{popped2 = }")
print(f"{my_list = }")
```

```python
# sort
a = [1, 2, 5, 1, 10, 42]
print(f"{a = }")
a.sort()
print(f"{a = }")

print("\n")
b = ["2", "1", "3", "4", "7", "3"]
print(f"{b = }")
b.sort()
print(f"{b = }")

ords = [ord(symb) for symb in b]
print(f"{ords = }")

print("\n")
c = ["asd", "dasdadasd", "aa", "adsda", "dasd"]
print(f"{c = }")
c.sort()
print(f"Default sort: {c = }")
c.sort(key=len)
print(f"Sort with `len` key: {c = }")
```

Зверніть увагу - коли ви користуєтесь методами, які змінюють структуру ліста (`append`, `extend`, `insert` `sort`), вам не треба робити переназначання його змінної, в той час коли вам треба щось отримати від ліста - `count`, `pop`, тоді треба ставити змінну.

## Словарі | Dictionaries

В списку - отримати той чи інший елемент можна за допомогою індексу, тобто `list[0] = elem`, інколи, хочеться замість індексу мати якийсь унікальний індекс, за допомогою якого можна отримати той чи інший елемент. Для цього використовують тип даних `dict`, скорочено від dictionary. Синтекст створення словаря - викорситання фігурних скобок, `{` з перечисленням ключ-значень через символ `:`, `key: value` який має закінчитися фігурною скобкою - `}`

```python
key = "Name"
grade = 5
my_dict = {key: grade}
print(f"{my_dict = }")

my_dict = {'Max': 12}
print(f"{my_dict = }")
```

```python
empty_dict = {}
print(empty_dict)
```

Елементами словарів є пари - ключ (key), значення (value). Ключі мають бути унікальними, в той час як значення - ні.

```python
key1 = "russia"
key2 = "Ukraine"

value1 = value2 = "Kyiv"

capitals = {key1: value1, key2: value2}
print(f"{capitals = }")
```

Як ключ, так і значення можут бути різних типів даних, `int`, `bool`, `str`, `list`, але на значення ключів є певні обмеження - так наприклад ви не можете створити ключ з типа даних, від якого не можна отримати кеш (наприклад - `list`)

```python
qwe = {True: "Something", False: "Something again", 1: [1, 2, 3], "Zieg": "Heil!"}
print(f"{qwe = }")
```

Альтернативою створення листів, як і з `list` є виористання типа даних як функції, тобто `dict()`

```python
empty_dict = dict()

print(f"{empty_dict = }")

dictionary = dict(key1 = "value", key2 = 'value')
print(f"{dictionary = }")

dictionary = dict([(1, True), (5, False)])
print(f"{dictionary = }")
```

Отримати значення від словника можна за допомогою синтаксиса, схожого на отримання значення за індексом `dict[key]`, та за допомогою методу `dict.get(key)`

```python
dicty = {3213: "User1", 1134: "User2"}

key = 3213
print(f"{dicty[key] = }")

print(f"{dicty.get(key) = }")
```

У випадку, коли такого індекса немає - інтерпритатор підіймає `KeyError`:

```python
dicty = {3213: "User1", 1134: "User2"}

key = 2235
print(f"{dicty[key] = }")
```

Додати об'єкт до словаря можно за допомогою наступного синтаксису:

```python
dicty[1111] = "User3"
print(dicty)
```

Або користуючись методом `dict.update(key = value)`

```python
dicty.update({5341: "User4"})
print(dicty)
```

## Кортежи | Tuple

Кортеж це тип даних який по факту є **списком що неможливо змінити**, тобто після його створення, змінити якісь елементи, додати в початок, або кінець списку неможливо. Має тип даних `tuple`. Цей тип даних корисен тим, що на відміну від `list`, займає набагато меньше пам'яті та є своєрідним "захистом від дурака"

Ситнаксис створення - аналогічен `list`, проте з використанням круглих дужок замість квадратних - перелік елементів ведеться також через кому.

```python
empty_tuple = ()
print(f"{type(empty_tuple) = }")

my_tuple = (1, 2, 3, "a", "dude", "doh", True, [[[[5]]]])
print(f"{my_tuple = }")
```

Іншим варіантом створення `tuple` є, як вже прийнято, використання його типу данних як функції - `tuple()`:

```python
empty_tuple = tuple()
print(f"{type(empty_tuple) = }")
my_tuple = tuple([1, 2, 3, "a", "dude", "doh", True, [[[[5]]]]])
print(f"{my_tuple = }")
```

Також, `tuple` з одним елементом має створюватися наступним чином:

```python
not_tuple = ('3')
print(f"{not_tuple = }")
print(f"{type(not_tuple) = }")

definetly_a_tuple = ('3', )
print(f"{definetly_a_tuple = }")
print(f"{type(definetly_a_tuple) = }")
```

Зверніть увагу - що викорастання типу даних як функції створює з іншого типу даних (`list`) тип даних `tuple`, по факту, використання будь якого типу даних як функції - намагання змінити тип даних з будь якого на другий. Але перетворення можливе лише за виконання вимог до того типу даних, який має бути перетворений

З іншого - `tuple` підтримує всі ті самі функції, що можуть бути застосовані до `list` (та й будь якої послідовності) типу `len`, `all`, `max`, `bool` і т.д. 

Щодо методів - `tuple` має такіж методи, що й `list`, однак тількі ті - що не змінюють сам `tuple`, тобто наприклад, ви **не можете** зробити `tuple.append(value)` бо цей тип даних є незмінюємим.

### Множини & Статичні Множини | Set & FrozenSet

Множини `set` та статичні множини `frozeset` - це тип даних в python, який тісно пересекається з множинами з алгебри та початку аналіза: [множини wiki](https://uk.wikipedia.org/wiki/%D0%9C%D0%BD%D0%BE%D0%B6%D0%B8%D0%BD%D0%B0).

Теорія множин дуже важлива в комп'ютерних науках і, зокрема, в криптографії. 

В пайтон множини являють собою типи даних `set` та `frozenset`, ці типи даних це "контейнери", які містять в собі лише **унікальні** елементи. Різниця між `set` та `frozenset` в першому наближенні такаж сама як і різниця між `list` та `tuple`. `set` - можливо змінити, `frozenset` 

Створюється `set` за допомогою фігурних дужок `{}`, де елементи `set` перечіслуються через кому, зверніть увагу, що користвуючись таким типом створення, неможливо створити пустий `set`:

```python
my_set = {1, 2, 3, 4, 5}
print(f"{my_set = }")

my_set2 = {"a", "a", "a", "b", "b", "b"}
print(f"{my_set2 = }")
```

Для створення пустої множини та статичної множини використовують типи даних як функції - `set()` та `frozenset()`

```python
empty_set = set()
print(f"{empty_set = }")
print(f"{type(empty_set) = }")

empty_frozenset = frozenset()
print(f"{empty_frozenset = }")
print(f"{type(empty_frozenset) = }")

set_w_elems = set([1, 2, 3])
print(f"{set_w_elems = }")

set_w_elems = set("abcd")
print(f"{set_w_elems = }")

frset_w_elems = frozenset([1, 2, 3])
print(f"{frset_w_elems = }")

frset_w_elems = frozenset("abcd")
print(f"{frset_w_elems = }")
```

Виходячі з алгоритмів, за допомогою яких створена така структура даних як `set`, цей тип даних є послідовністю, але не упорядкованою, кожен раз, коли ви викликаєте `set`, елементи всередині знаходяться в випадковому порядку. Тому `set` не підтримує індексацію, зрізі і т.п.

Якщо вам треба спочатку створити сет, а потім отримувати елементи поіндексово, створіть з сету `list`, за допомогою використання цього типу даних як функції, тобто `list(my_set)`

```python
my_set = set("asdkas;dk")
my_list = list(my_set)
my_list
```

Множини підтримують ті ж самі `built-in` функції, які підтримують послідовності. Також працюють з операторами `in`, `==`, `>` і т.д. 

Окремим пунктом хотілося-б виділити основний функціонал цієї структури даних, а саме імплементації алгебраічних операцій над множинами:

Імплементуємо математичні операції над множинами за допомогою інструментарію `set`:

```python
try:
    import matplotlib.pyplot as plt
    from matplotlib_venn import venn2, venn3
except ModuleNotFoundError:
    !pip install matplotlib-venn
```

```python
set1 = set(['A', 'B', 'C', 'D', "K", "T"])
set2 = set(['D', 'E', 'F', "A", "P", "D"])

venn2([set1, set2], ('set1', 'set2'))
```

На графіку вище ви бачите діаграму ейлера-вейна, яка показує кількість множини `set1`, `set2` та їх перетин. Цифрами показана кількість елементів в різних під-множинах, які можуть бути сформовані завдяки операціям з множинами:

```python
# Difference

print("Set 1 \n")
difference_subset1 = set1.difference(set2)
print(f"{difference_subset1 = }")
print(f"{len(difference_subset1) = }")

print("\nSet2 \n")
difference_subset2 = set2.difference(set1)
print(f"{difference_subset2 = }")
print(f"{len(difference_subset2) = }")
```

```python
# Intersection

print("Set 1 \n")
intersection_subset1 = set1.intersection(set2)
print(f"{intersection_subset1 = }")
print(f"{len(intersection_subset1) = }")

print("\nSet2 \n")
intersection_subset2 = set2.intersection(set1)
print(f"{intersection_subset2 = }")
print(f"{len(intersection_subset2) = }")
```

```python
# Symmetric Difference

print("Set 1 \n")
sym_difference_subset1 = set1.symmetric_difference(set2)
print(f"{sym_difference_subset1 = }")
print(f"{len(sym_difference_subset1) = }")

print("\nSet2 \n")
sym_difference_subset2 = set2.symmetric_difference(set1)
print(f"{sym_difference_subset2 = }")
print(f"{len(sym_difference_subset2) = }")
```

Як вже було сказано `set` це змінюємий тип даних, а `frozenset` - ні, тому всі методи, які змінюють структуру, елементи всередині, тобто до `set` має метод `add`, в той час як `frozenset` - ні.
