Між тим щоб запустити будь-яку програму, ваш комп'ютер виконує велику кількість дій. В фізичному плані - процесор це лише величезна кількість транзістроів. Але з об'єднанням цих транзисторів в логічні схеми, логічні гейти, та додавши до цього інструкції, рівні абстракції, пам'ять і т.д. він претворюється на те за чим ви працюєте зараз.

Для того, щоб створити машиний код - використовуються **Компілятори** та **Інтерпритатори**. Як компілятор так і інтерпритатор конвертують інструкції однієї мови програмування в іншу. Найчастіше їх використовують для конвертації інструкцій високорівневої мови програмування (C, Python, Java, e.t.c) в бінарну форму, або машинний код.

Виходячи з типу того, яким чином з інструкцій верхнього рівня створюються інструкції для обробки процесором - розділяють мови прокраммування - compled languages / interpreted languages.

Різниця між тим, як працюють (в першому наближенні) компілятор та інтерпритатор - компілятор транслює високорівневу мову повністю, тобто після роботи компілятора з'являться файл .bin який є файлом машинного коду і запускається теж повністю. В свою чергу інтерпритатор створює проміжну мову, кожна інструкція якого виконується окремо та послідовно. 

Python - це interpreted language. Як і люба мова програмування що інтерпритується - пайтон транслює інструкції, написані в файлі програми в байт-код, який потім послідовно запускається в віртуальній машині - PVM

При опрацюванні кода пайтон виконує три умовні кроки:
* Ініціалізація
* Компіляція
* Виконання

**Ініціалізація**: інтерпритатор считує код, перевіряючи його форматування та синтаксіс, у разі помилки у цих двух крітеріях - викликається помилка.

**Компіляція**: інтерипритатор переводить інструкції програми в байт-код (не машинний код), байткод - низькорівневий вид ваших інструкцій

**Виконання**: Після створення байткоду, байткод поетапно відправляється на виконная в PVM, якщо під час виконання такого коду з'явилася помилка, яка не була оброблена - виконання завершується з виводом отриманої помилки

Завдяки такій реалізації код написаний на пайтоні, у разі якщо не було допущено синтаксичних помилок, якась частина кода, яка була запущена може бути виконана до того, як отримається помилка - наприклад

```python
test1 = 5 + 5
print(test1)
test2 = 5 / 0
```

## Змінні

Змінна - (variable, var) це 'контейнери' для присвоєння та збереження даних, створення змінної називається її об'явленням (declaring, defining), в python - змінні об'являються наступним чином:

```python
answer = 42
```

Завдяки оператору `=` (який називається оператор присвоєння) (assignment operator), ми створили змінну з ім'ям `answer` та наділили її значенням `42`. Ми можемо викликати цю змінну користуючись командою `print` та подивитись її значення:

```python
print(answer)
```

Правила створення змінних наступні:
* вони не мають починатись з числа do: `grade_5 = 5` don't `5grade = 5`
* ім'я має бути написано без пробелів do `grade_5 = 5` don't `grade 5 = 5`
* не дозволено використовувати `:'",<>/?|\!@#%^&*~-+`
* найліпша практика створення змінної - нижній регістр зі знаками `_` в якості пробелів
do: `my_long_variable_name` dont: `MyLong_VariableName`

Приклад можливостей пайтон в створенні змінних:

```python
my_dog_age = 5
print(my_dog_age)

my_dog_age = 12
print(my_dog_age)

my_dog_age = 12 + 5
print(my_dog_age)

my_age = my_dog_age
print(my_age, my_dog_age)

my_dog_age = my_dog_age + 5
print(my_dog_age)

my_age = my_girlfriend_age = 32
print(my_age, my_girlfriend_age)

my_age, my_sister_age, my_girfriend_age = 32, 16, 16
print(my_age, my_sister_age, my_girfriend_age)
```

Коли ми присваюємо якійсь змінній будь яке значення, насправді ми даємо цій змінній ссилку на об'єкт, на відміну від деяких інших мов програмування, наприклад С, коли змінній прісвоюється значення індексу в пам'яті комп'ютера. 

Що це означає - коли ми використвуємо операцію присвоєння - спочатку оброблюється права частина виразу, а потім - ліва. Спершу в пам'яті ствоюється об'єкт - наприклад `300`, далі ссилка на цей об'єкт присвоюється новоствореній змінній

```python
my_var = 300 # спочатку створюється 300
my_var = 3 + 5 # в даному випадку - спочатку оброблюється вираз '3 + 5'
my_var = my_var2 = my_var3 = 300 # всі змінні отримали туж саму ссилку
```

Щоб побачити - що ці змінні отримали ссилку на той самий об'єкт, звернемося до індексу в пам'яті у всіх цих об'єктів за допомогою команди `id`

```python
print(id(my_var), id(my_var2), id(my_var3))
```

Якщо ми створемо декілька об'єктів в правій частині, а потім 'розкинемо' їх по змінним, нічого не зміниться, бо пайтон вже створив ці об'єкти, і поки він бачить що вони будуть використовуватися далі, він буде давати ту ж саму ссилку.

```python
my_var, my_var2, my_var3 = 300, 300, 300 
print(id(my_var), id(my_var2), id(my_var3))
```

Але якщо в даній строчці - об'єкт більше не буде використовуватися, але буде використовуватися в наступній, змінна отримає інше значення `id`

```python
my_var = 300
my_var2 = 300
print(id(my_var), id(my_var2))
```

Це є справедливим для більшості можливих введених даних, окрім декількох випадків, наприклад для чисел `[-5: 256]`

```python
m = 5
n = 5
print(id(m), id(n))
```

Також у рамках теми змінних треба коснутися **Типізації**. Типізацією в першому наближенні описується те як поводять себе дані в тих чи інших умовах, та тим як ці дані створюються та записуються в пам'ять.

## Оператори 

Оператори використовуються для УВАГА! операцій над даними та значеннями.
Розділяють наступні категорії операторів:

* Оператори аріфметики (Arithmetic operators)
* Оператори присвоєння (Assignment operators)
* Оператори порівняння (Comparison operators)
* Оператори логіки (Logical operators)
* Оператори ідентифікації (Identity operators)
* Оператори членства (Membership operators)
* Оператори побітових операцій (Bitwise operators)

### Оператори аріфметики

```python
5 + 5 # addition
5 - 5 # substraction
5 * 5 # multiplication
5 / 5 # division
5 % 5 # modulus
5 ** 5 # exponention
5 // 5 # floor division
```

Може здатися логічним - що аріфметичні оператори можуть бути використані тількі з чисельними типами, але це не так, деякі типи даних також підтримують роботу з операторами (чому - буде пояснено в лекціях з OOP). Наприклад - роботу з деякими операторами підтримують строки `str`

```python
print("a" * 10)
print("a" + "a")
```

### Оператори присвоєння

```python
a = 5 # assignment

# arithmnetic assignment
a += 5 # means a = a + 5
a -= 5 # means a = a - 5
a *= 5
a /=5
a %= 5
a //= 5
a ** 5

# bitwise assignments (won't be discussed)
a = 5
a &= 5
a |= 5
a ^= 5
a >>=5
a <<=5
```

### Оператори порівняння

```python
a == 5 # equal
a != 5 # not equal
a > 5 # greater than
a < 5 # less than
a >= 5 # greater than or equal to
a <= 5 # less than or equal to
```

Зверніть увагу - результатом операції порівнняння є новий для вас тип даних - `bool`, який має рівно два значення - `True` або `False`, детальніше цей тип даних буде розібраний на лабораторній роботі.

### Оператори логіки

```python
a = True
a and False # Returns True if both statements are true
a or False # Returns True if one of the statements is true
not a # Reverse the result, returns False if the result is true
```

Але ці оператори не є настількі простими як це здається, наприклад оператори також можуть працювати усіма типами даних, та їх поведінка не завжди очевидна, але зрозуміла при детальному розборі взаємодії даної операції з визначеним типом даних.

```python
print(1 and 5)
print(0 and 5)
print(1 or 5)
print(5 or 1)
print(0 or 5)
print(0 and 0)
print(not 5)
print(not 0)
```

### Оператори членства

```python
a = "o"
a in "oOo" # Returns True if a sequence with the specified value is present in the object
a not in "oOo" # Just a logic NOT applied to the operator IN

print("a" in "Alphabet")
print("a" in "Python")
print("a" not in "Python")
print("p" in "Python")
print("P" in "Python")
```
