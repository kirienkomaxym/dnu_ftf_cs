## Строки

Строки це тип даних який використовується для маніпуляцій з текстом. Строки можуть вміщати в собі букви абетки, цифри, та спеціальні симболи. В python є вбудована підтримка строк, яка репрезентована класом `str`. Строки в python, як і в інших мовах програмування репрезентують послідовність символів **Unicode**. **Unicode** це система, розроблена для того, щоб відображати символи з натуральних мов. В Unicode - кожна літера або символ репрезентована унікальною комбінацією з 2х байтів. 

Як можна побачити з табиці вище - кожен символ має свій унікальний нумер, репрезентований двома байтами, щоб отримати унікальний номер того чи іншого строкового символа - можна скористатись функцією `ord`, та передати їй в якості аргумента наш символ. Для отримання символу по нумеру - можна скористатись функцією `chr`, та передати їй в яксоті аргумента число в типі `int`.

```python
my_char = "a"
my_char_ord = ord(my_char)
print("My char number in decimal:", my_char_ord)
print("My char number in bytes:", bin(my_char_ord))
my_char_from_ord = chr(my_char_ord)
print("My char from decimal:", my_char_from_ord)
```

Інколи - нам буде необхідно форматувати строки, для цього існують наступні механізми:
* Форматування з використанням оператора '%'
* Форматування з темплейтів
* Використання f-строк

Oператор '%' - встроєний оператор, коли він взаємодіе зі строками, він називається оператор інтерполяції, користучись цим оператором для форматування - треба передати тип даних, яким буде форматована строка.

```python
print("I bought %d Euro worth of %s!" %(200, 'cookies'))
print("My mean grade is %d, probably, because Im %s" %(3, 'lazy'))
```

При використанні форматування з темплейтів - використовується наступний сінтаксіс. Спершу, ми створюємо строку, залишаючи в ній символи `{}`, або вписуємо в ці фігурні скобки майбутнє ім'я змінної, через яку ми потім передамо значення для форматування:

```python
my_template = "{} was shot {} times"
template_formatted = my_template.format('50cent', '5')
print(template_formatted)

# it actualy works as a oneliner
my_template = "{} was shot {} times".format('2pac', 9)
print(my_template)

# oh! not reusable!
print(my_template.format('Biggie', 3))

template_with_keywords = "{rapper} was shot {num} times"
print(template_with_keywords.format(rapper="Tentacion", num="1"))

# # keyword template does not support not providing the keys,
# # and args template doesnt support keywords
# print(template_with_keywords.format("Tentacion", "1"))
```

Останнім способом форматування є використання f-string, на теперишній час воно використовується найчастіше.

```python
num = 3
rapper_name = 'Biggie'
print(f"{rapper_name} was shot {num} times!")

# if you want to provide non-variable values it works too:
print(f"{'Поплавський'} was shot {0} times!")

# in newer python version, it also supports printing the variables name, e.g.
print(f"{rapper_name=} was shot {num=} times!")
```

Як вже було сказано, строки підтримують можливість використання операторів, наприклад:

Також, строки підтримують деякі вбудовані функції python, такі як `len` - визначення кількіості елементів, а також функції переводу строки в інший тип даних, наприклад - `int`.

```python
string = "aa bb cc"
print(len(string))

string = ""
print(len(string))

string_with_int = "100000"
print(int(string_with_int))
```

Ще одним важними елементом сінтаксісу пайтон є індексація та зрізи (indexation and slicing). Ці інструменти працюють на всіх типах данних, як являються індексуємими, тобто будь якими послідовностями, з іншими типами даних, що індексуються, ви ознайомитесь на наступних лекціях та лабах.

Індексація можливість отримати елемент з послідовності опираючись на його індекс. Синтаксіс отримання елементу по індексу - `sequence[index]`, де в `[]` вказується індекс елементу.

Зрізи - можливість отримати з будь якої послідовності іншу послідовність, в залежності від вказаних індексів та шагу послідовності, синтаксис наступний - `sequence[start:stop:step]`, де `start` - індекс початку зріза, `stop` - індекс кінця зріза, `step` - шаг зріза.

```python
string = "My name is Taras"
print(f"First elem {string[0]}") # indexes starts with 0, ends with len(obj) - 1
print(f"Length of our sequence: {len(string)}")
print(f"Last elem of our sequence {string[-1]}")

string = "Yet another testing string"
third_elem = string[3]
print(f"{third_elem=}")

mean_elem = string[int(len(string) / 2)]
print(f"{mean_elem=}")

reverse_third_elem = string[-3]
print(f"{reverse_third_elem=}")

# # note that len(sequence) gives you amount of items in sequence
# # so in case of 3 element sequece 'abc', len(`abc`) == 3
# # while last element `c` have either index 2 and -1
# abc = "abc"
# print(abc[len(abc)])
```

Також, завдяки індексації ми можемо присвоювати значенням послідовності нові значення:

```python
string = "What?! My name is - Slim Shady!"
my_slice = string[0:3]  # get three elements from index 0 to index 3
print(f"{my_slice=}")

full_slice = string[:] # note that none of the `start` `stop` `step` is actually required, but operator `:`
print(f"{full_slice=}")

another_full_slice = string[::]  # result will be the same
print(f"{another_full_slice=}")

slice_from_third = string[3:]  # you can start from the third
print(f"{slice_from_third=}")

slice_from_negative_third = string[-3:] # works in both directions
print(f"{slice_from_negative_third=}")

slice_untill_seventh = string[:7]
print(f"{slice_untill_seventh=}")

slice_until_negative_seventh = string[:-7]
print(f"{slice_until_negative_seventh=}")

reversed_slice = string[::-1]
print(f"{reversed_slice=}")

negative_index_slice = string[-7:-3]
print(f"{negative_index_slice=}")

every_second_slice = string[::2]
print(f"{every_second_slice=}")

negative_index_reversed_slice = string[-3:-7:-1]
print(f"{negative_index_reversed_slice=}")
```

```python
# another approach of making slices is by using `slice` built-in, 
# actualy it works the same, unless you want to store some of 
# the slices for further usage

string = "Бібадох"
my_slice = slice(4, 7)
string[my_slice]
```

Методи, на відміну від функцій, викликаються разом з об'єктом - синтаксис виклику метода - `object.method`. Кожен тип даних обладає своїми методами, в контексті розмови про об'єкти `str` ви вже бачили метод `str.format`, детальніше про методи ви дізнаєтесь з лекцій по OOP.

```python
string_w_uppercase = "SomeString"
lowercased = string.lower()
print(lowercased)
```

Скорестаємося деякими з цих методів:

```python
sentence = "I am Student"
first_word, second_word, third_word = sentence.split()
print(f"{first_word=}, {second_word=}, {third_word=}")

replaced_sentence = sentence.replace("Student", "Teacher")
print(f"{replaced_sentence=}")

I_index = sentence.find("I")
print(f"{I_index=}")

i_index = sentence.find('i')
print(f"{i_index=}")  # -1 means was not found

# also they can be combined
replaced_sentence = sentence.lower().replace('i', 'God Bless')
print(f"{replaced_sentence}")
```

## Числа

Для збереження та обробки чисел в пайтон існує декілька типів даних:
* int цілі числа `[-inf, inf]`
* float числа з плаваючою точкою `[-inf., inf.]`
* complex комплексні числа `[-inf + infj, inf + infj]`

Для створення чисел використовується той синтаксіс:

```python
integer = 5
print(f"{type(integer)=}")

float_point = 5.
print(f"{type(float_point)=}")
float_point = 5e3
float_point = 5e-3

complex_point = 5e-3 + 5j
print(f"{type(complex_point)=}")
```

Всі ці типи даних підтримують роботу з усіма аріфметичними і бітовими операціями.
Також в них є свої методи, як це було з класом `str`.
До нашого списку типів даних додалися наступні типи - `float`, `complex`, ці типи даних також мають свої функції з тою самою назвою що і тип даних, викликаючи котрі можна привести аргумент (якщо це можливо) до вказаного типу, наприклад:

```python
integer = 5
floater = float(integer)
print(f"{floater=}")

integer = int(floater) # works vise versa
print(f"{integer=}")

complx = complex(integer)
print(f"{complx=}")

# integer = int(complx) # except for complex
# print(f"{integer=}")

# floater = float(complx) # it cant be converted
```

## Boolean

Булєві значення як вже було вказано раніше це значення `True`, `False`, які використовуються для двоїчної логіки, перевірок, також вони є результатом виконання операторів порівнняння, e.g. `==`, `>`, `<`. Вони обладають власним типом даних - `bool`, користюючись якою, можна привести інший тип даних до типу даних `bool`

```python
a = True
print(f"{type(a)=}")
b = False
print(f"{type(b)=}")

a = 5
print(f"{bool(a)=}")

a = 0
print(bool(a))

print(bool("abc"))

print(bool(""))

print(bool(0.))

print(bool(0.0001))
```

Також бульові оператори можуть взаємодіяти з операторами `not`, `and` `or` і тп.

```python
res = not True

res2 = True and False

res3 = True or False

res4 = True or not False
```
