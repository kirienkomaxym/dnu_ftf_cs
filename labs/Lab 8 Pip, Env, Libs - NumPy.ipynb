## NumPy

Ви напевно знайомі з поняттям векторів, матриць, а можливо й тензорів (в першому наближенні це багатовимірні матриці). 

`numpy` це бібліотека, для швидкої та ефективної роботи з масивами (arrays). Ці масиви - можуть бути як одновимірні (вектори), двовимерні (матриці), багатовимерні (тензори), але в парадігмі `numpy`, вони є масивами.

Звернемося до офіційної документації `numpy`, і подивимося на її опис:

```
NumPy is the fundamental package for scientific computing in Python. It is a Python library that provides a multidimensional array object, various derived objects (such as masked arrays and matrices), and an assortment of routines for fast operations on arrays, including mathematical, logical, shape manipulation, sorting, selecting, I/O, discrete Fourier transforms, basic linear algebra, basic statistical operations, random simulation and much more.
```

Давайте приблизно роздивимося базовий функціонал, пройдемося по модулям, які пропонує `numpy`:

Центральним об'єктом, яким ми оперуємо при використанні цього пакету є масив (array), типа даних `numpy.ndarray`. 

Масив — це сітка значень, яка містить інформацію про необроблені дані, способи пошуку елемента та способи інтерпретації елемента. Він має сітку елементів, які можна індексувати різними способами. Усі елементи одного типу, називаються **dtype** масива.

Масив може бути проіндексований кортежем невід’ємних цілих чисел, логічними значеннями, іншим масивом або цілими числами. **Ранг** масиву - це кількість вимірів. **Форма** масиву — це кортеж із цілих чисел, що визначає розмір масиву вздовж кожного **виміру**.

````python
import numpy as np
````

Створемо простий 1D масив, або вектор, в математичній нотації:

````python
my_list = [1, 2, 3, 4, 5]
my_array = np.array(my_list)
````

Як було зазначено вище, основними параметрами масива є кількість **вимірів** (dimensions) або **ранг**, **розмір** масива - або кількість елементів всередині (size), та **структура** або форма масива (shape). І також тип елемента всередині цього масива:

Кожний інстанс np.ndarray має ці параметри, і звернутися до них можна завдяки його атрибутам - `<array>.shape`, `<array>.ndim`, `<array.size>`

````python
my_array.shape
````

````python
my_array.ndim
````

````python
my_array.size
````

````python
my_array.dtype
````

Також, у вас є можливість отримувати значення з масива за індексом, та отримувати зрізи від цього масива:

````python
my_array[0]
````

````python
my_array[2:5]
````

Як вже було сказано, нампай підтримує можливість оперувати багатовимірними масивами, проте як вони виглядають?

Приблизно так - і тут додається ще одна концепція - концепція вісей (axes), цого масива, завдяки якій, ми можемо оперувати зрізами і індексацією елементів всередині масива.

Як можна побачити з картинки, індексація вісей починається з `0`, так само, якби ми робили вложені лісти, пам'ятаєте ці задачі?

### broadcasting

Уявимо що ви програмуєте гру, хай це буде 2д платформер, і цей платформер оперує в двох "плоскостях" або вимірах, тобто висота `y` та довжина `x`. В якийсь момент часу, вам необхідно визначити швидкість персонажа, можливо, з фізики ви знаєте що швидкість визначається по кожній вісі, і є вектором:

````python
speed_vector = np.array([0, 4])
````

````python
speed_vector
````

Тобто, уявимо що перший індекс відповідає за швидкість по х, а другий - за швидкість по у, і вам необхідно перевести швидкість км\год на мілі\год. Як варіант - ви можете скористатись створенням масива того ж розміру та помножити їх:
